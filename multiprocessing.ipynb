```python
from multiprocessing import Process
from time import time,sleep
```

```python
class Exponents(Process):
    def __init__(self,x,y):
        Process.__init__(self) # witthout calling this it will cause runtime error of Thread.__init__() not called
        self.x=x
        self.y=y
        
    def run(self):
        step=1 if self.y<0 else -1
        # range(-5) wont work use range(-5,0), but range(5,0) also wont work use range(5,0,-1)
        for i in range(self.y,0,step):
            print(self.x**i)
            #sleep(0.2)

        
```

```python
x=10
y=50
exponent_obj1=Exponents(x,y)
exponent_obj2=Exponents(x,-y)
```

```python
exponent_obj1
```

```python
ts=time()
exponent_obj1.start()
exponent_obj2.start()

exponent_obj1.join()
exponent_obj2.join()
te=time()
print("Multiprocessing took {} s".format(te-ts))
```
