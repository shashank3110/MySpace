```python
import numpy as np
```

```python
a=1/100
n=2
u=1#myu
x=np.array((0.5,0.5))

```

```python
def get_g(x):
    xt=np.transpose(x)

    g1=np.dot(xt,x)-1
    g2=-x[0]
    return g1,g2
```

```python
def get_dg(x):
    dg1=[2*x[0],2*x[1]]
    dg2=[-1,0]
    return dg1,dg2
```

```python
def get_fx(x):
    f=0
    for i in range(n):
        f+=x[i]
    return f
```

```python
def get_dfx():
    return np.array((1,1))
```

```python
def compute_gsum(x):
    
    g1,g2=get_g(x)
    dg1,dg2=get_dg(x)
    sum=np.empty((1,2))
    sum[0][0]=dg1[0]/g1 + dg2[0]/g2
    sum[0][1]=dg1[1]/g1 + dg2[1]/g2
    
    return sum[0]
```

```python
def log_barrier(x,u):
    sum=0
    g1,g2=get_g(x)
    dg1,dg2=get_dg(x)
    print("Gs",g1,g2)
    sum=np.log(-g1)+np.log(-g2)
    F=get_fx(x)-u*sum
    return F
```

```python
def compute_log_barrier_gradient(x):
    dfx=get_dfx()
    iter=10
    u=1 #myu
    a=0.01#alpha
    while iter>=0:
        DF=dfx-u*compute_gsum(x)
        x,alpha=compute_grad_descent(x,a,DF,u)
        a=alpha
        xp=x
        u=u/2
        iter=iter-1
    return xp
```

```python
def compute_grad_descent(x,a,DF,u):

    print(np.shape(DF))
    delta=-(DF/np.linalg.norm(DF))
    theta=1
    w=0.01
    w1=1.2
    w2=0.5
    delmax=np.inf
    dfx=get_dfx()
    i=0
    
    #func=get_fx
    func=log_barrier
    while  np.linalg.norm(a*delta) <theta :

        print("alpha=",a)
        
        
        delta=-(DF/np.linalg.norm(DF))
        j=0
        print("DFT",np.transpose(DF))
        #print(np.dot(np.transpose(x),x)-1)
        #print( func(x) +w*(np.dot(np.transpose(DF),a*delta)))
        while func(x+a*delta,u) > func(x,u) +w*(np.dot(np.transpose(DF),a*delta)) :
                #num func evaluations
                a=w2*a
                j=j+1
                print("num of inner evaluations",j)
        print("num evaluations",i+1)
       
        
        x=x+a*delta
        DF=dfx-u*compute_gsum(x)
        print("x=",x)
        a=min(w1*a,delmax)
        i=i+1
        if i==10:
            break

    print("End of Gradient Descent alpha=",a)
    return x,a
```
