```python
from threading import *
from time import time,sleep
```

```python
class Exponents(Thread):
    def __init__(self,x,y):
        Thread.__init__(self) # witthout calling this it will cause runtime error of Thread.__init__() not called
        self.x=x
        self.y=y
        
    def run(self):
        step=1 if self.y<0 else -1
        # range(-5) wont work use range(-5,0), but range(5,0) also wont work use range(5,0,-1)
        for i in range(self.y,0,step):
            print(self.x**i)
            #sleep(0.2)

        
```

```python
x=10
y=50
exp1=Exponents(x,y)
exp2=Exponents(x,-y)
```

```python
ts=time()
exp1.start()
exp2.start()
exp1.join()
exp2.join()
te=time()
print("Threading took {} s".format(te-ts))
```
